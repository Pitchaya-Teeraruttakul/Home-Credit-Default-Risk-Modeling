```python
# Drop columns and rows

# DAYS_EMPLOYED: Replace 365243 (means "not employed") with NaN 
appTrain['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)

# FLAG_DOCUMENT_i (binary indicator like 0 and 1) which have 18 columns 
# --> Drop all if they provide little to no variance
flags = [col for col in appTrain.columns if 'FLAG_DOCUMENT_' in col]
low = [col for col in flags if appTrain[col].sum() < 300]
appTrain.drop(columns=low, inplace=True)

# Drop highly missing columns
missing_threshold = 0.4
missingF = appTrain.isnull().mean()
missingF = missingF[missingF > missing_threshold].index.tolist()
appTrain.drop(columns=missingF, inplace=True)

# Drop constant or near-constant features
conCol = [i for i in appTrain.columns if appTrain[i].nunique() <= 1]
appTrain.drop(columns=conCol, inplace=True)

nearConst = [
    i for i in appTrain.columns 
    if appTrain[i].value_counts(normalize=True, dropna=False).values[0] > 0.99
]
appTrain.drop(columns=nearConst, inplace=True)

# Correct unrealistic values
# Cap AMT_INCOME_TOTAL at 99th percentile
incomeCap = appTrain['AMT_INCOME_TOTAL'].quantile(0.99)

appTrain['IS_INCOME_CAPPED'] = (appTrain['AMT_INCOME_TOTAL'] > incomeCap).astype(int)
appTrain['AMT_INCOME_TOTAL'] = np.where(
    appTrain['AMT_INCOME_TOTAL'] > incomeCap,
    incomeCap,
    appTrain['AMT_INCOME_TOTAL']
)

# Cap CNT_CHILDREN at 5
appTrain['IS_CHILDREN_CAPPED'] = (appTrain['CNT_CHILDREN'] > 5).astype(int)
appTrain['CNT_CHILDREN'] = np.where(
    appTrain['CNT_CHILDREN'] > 5,
    5,
    appTrain['CNT_CHILDREN']
)

# Handle categorical variables
appTrain['CODE_GENDER'].replace('XNA', np.nan, inplace=True) 
# When a value is clearly invalid (XNA = unknown, error), we treat it as missing. It's consistent with IFRS 9 data quality standards
appTrain['NAME_INCOME_TYPE'].replace('Maternity leave', 'Working', inplace=True)
# Maternity leave appears in very few rows (100). These applicants are temporarily out of work but will return, so grouping into 'Working' avoids one-hot sparsity
appTrain['NAME_FAMILY_STATUS'].replace('Unknown', np.nan, inplace=True)
# Unknown is a placeholder. Marking as missing is better than pretending it's a real category

# Convert binary flags to categorical
binary_flags = ['FLAG_OWN_CAR', 'FLAG_OWN_REALTY']
for col in binary_flags:
    appTrain[col] = appTrain[col].map({'Y': 1, 'N': 0})

# REGION_RATING_CLIENT and REGION_RATING_CLIENT_W_CITY are near-duplicates
appTrain.drop(columns=['REGION_RATING_CLIENT_W_CITY'], inplace=True)

# group basic sanity flags --> create binary summary flags from the raw contact indicator
appTrain['HAS_PHONE'] = ((appTrain['FLAG_PHONE'] == 1) | (appTrain['FLAG_WORK_PHONE'] == 1)).astype(int)
appTrain['HAS_CONTACT_INFO'] = ((appTrain['FLAG_PHONE'] + appTrain['FLAG_EMAIL']) > 0).astype(int)
```

```python
installmentsPayments = pd.read_csv("../data/installments_payments.csv")
installmentsPayments.head()
```

```python
# checking for missing value and NA
missing = installmentsPayments.isnull().mean().sort_values(ascending=False).reset_index()
missing.columns = ['feature', 'missingRatio']
missing = missing[missing["missingRatio"]>0]
missing.style.background_gradient(cmap='Reds')
```

```python
# Calculate days paid late or early
installmentsPayments['PAYMENT_DELAY'] = installmentsPayments['DAYS_ENTRY_PAYMENT'] - installmentsPayments['DAYS_INSTALMENT']

# Amount paid minus required
installmentsPayments['PAYMENT_DIFF'] = installmentsPayments['AMT_PAYMENT'] - installmentsPayments['AMT_INSTALMENT']

# Payment ratio
installmentsPayments['PAYMENT_RATIO'] = installmentsPayments['AMT_PAYMENT'] / installmentsPayments['AMT_INSTALMENT']
installmentsPayments['PAYMENT_RATIO'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Late or on-time payment
installmentsPayments['IS_LATE'] = (installmentsPayments['PAYMENT_DELAY'] > 0).astype(int)
installmentsPayments['IS_EARLY'] = (installmentsPayments['PAYMENT_DELAY'] < 0).astype(int)

# Partial and overpayment
installmentsPayments['IS_UNDERPAY'] = (installmentsPayments['PAYMENT_DIFF'] < -10).astype(int)
installmentsPayments['IS_OVERPAY'] = (installmentsPayments['PAYMENT_DIFF'] > 10).astype(int)

# Missed payment flag (zero or missing actual payment)
installmentsPayments['IS_MISSED'] = (installmentsPayments['AMT_PAYMENT'] == 0).astype(int)
```

```python
installmentAgg = installmentsPayments.groupby('SK_ID_CURR').agg({
    'PAYMENT_DELAY': ['mean', 'max', 'min'],
    'PAYMENT_DIFF': ['mean', 'min', 'max'],
    'PAYMENT_RATIO': ['mean', 'std'],
    'IS_LATE': 'sum',
    'IS_EARLY': 'sum',
    'IS_OVERPAY': 'sum',
    'IS_UNDERPAY': 'sum',
    'IS_MISSED': 'sum',
    'NUM_INSTALMENT_NUMBER': 'count',  # Total number of payments
})

installmentAgg.columns = ['INSTM_' + '_'.join(col).upper() for col in installmentAgg.columns]
installmentAgg.reset_index(inplace=True)
```

```python
display(installmentAgg.shape)
display(installmentAgg.columns.tolist())
```

```python
# bureauBalAgg need to merge to bureau first then merge to bureauAgg --> bureauBalAgg merge with SK_ID_BUREAU
# I can't use the bureauAgg --> features will redundant, therefore, need to merge bureauBalAgg to bureau
# then make aggregation again, so, we gonna get bureauAgg with bureauBalAgg info contained
bureauFull = bureau.merge(bureauBalAgg, how='left', on='SK_ID_BUREAU')
```

```python
display(bureauFull.shape)
display(bureauFull.columns.tolist())
```

```python
# aggregate bureauFull same way as I done before but need to add BB_ aggregation too
# avoid the redundant aggregation like mean of the mean --> but the others aggregation might sometime meaningful but better to do
bureauAgg = bureauFull.groupby('SK_ID_CURR').agg({
    'SK_ID_BUREAU': 'count',
    'DAYS_CREDIT': ['min', 'max', 'mean'],
    'DAYS_CREDIT_ENDDATE': 'mean',
    'AMT_CREDIT_SUM': ['sum', 'mean'],
    'AMT_CREDIT_SUM_DEBT': ['sum', 'mean'],
    'AMT_CREDIT_SUM_OVERDUE': ['sum', 'mean'],
    'DEBT_CREDIT_RATIO': 'mean',
    'IS_ACTIVE': 'sum',
    'IS_CLOSED': 'sum',
    'IS_SOLD': 'sum',
    'IS_BAD_DEBT': 'sum',
    'IS_REVOLVING': 'sum',
    'IS_CONSUMER': 'sum',
    'IS_MORTGAGE': 'sum',

    # Bureau Balance features (from bureauBalAgg)
    'BB_MONTHS_BALANCE_MIN': 'min',
    'BB_MONTHS_BALANCE_MAX': 'max',
    'BB_MONTHS_BALANCE_COUNT': ['mean', 'sum'], # add sum --> might have more meaningful to the model
    'BB_IS_LATE_SUM': 'sum',
    'BB_IS_DPD_30_SUM': 'sum',
    'BB_IS_DPD_60_SUM': 'sum',
    'BB_IS_DPD_120_SUM': 'sum',
    'BB_IS_CLOSED_SUM': 'sum',
    'BB_IS_NODATA_SUM': 'sum'
})

bureauAgg.columns = ['BUREAU_' + '_'.join(col).upper() for col in bureauAgg.columns]
bureauAgg.reset_index(inplace=True)
```

```python
# got the same shape as bureau.shape[1] plus bureauBalAgg.shape[1]
display(bureauAgg.shape)
display(bureauAgg.columns.tolist())
```

```python
# Join all the cleaned table --> now we can join all the cleaned table
final = (
    appTrain
    .merge(bureauAgg, on='SK_ID_CURR', how='left')
    .merge(installmentAgg, on='SK_ID_CURR', how='left')
)

print(final.shape)
# checking for the final table missing value (which features have the most missing)
print(final.isna().mean().sort_values(ascending=False).head(10)) 
# condider to drop the features that missing value > 50%
```

```python
# these features that have the number of missing value > 0.5 --> let's check, Can I drop it
final[['SK_ID_CURR','BUREAU_BB_MONTHS_BALANCE_COUNT_MEAN','BUREAU_BB_MONTHS_BALANCE_MIN_MIN','BUREAU_BB_MONTHS_BALANCE_MAX_MAX' ]].head(10)
```

```python
# Drop all columns with > 50% missing values
missing = final.isnull().mean()
final.drop(columns=missing[missing > 0.7].index, inplace=True)
```
